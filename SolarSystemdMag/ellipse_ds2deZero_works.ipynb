```python
import sympy as sp
import numpy as np
```

```python
xe, x, ye, y = sp.symbols('x_e, x, y_e, y ')
a, b, e = sp.symbols('a b e')
```

equation for ellipse

```python
# solving xe**2/a**2 + ye**2/b**2 = 1 for ye
ellipse_eqn = xe**2/a**2 + ye**2/b**2 - 1
```

solve for ye(xe)

```python
ye_xe_eqn = sp.solve(ellipse_eqn,ye)[1] #only 1 should be used since ye will be strictly positive if x and y are strictly positive
```

```python
ye_xe_eqn
```

the official 's' equation

```python
s_eqn = sp.sqrt((xe - x)**2 + (ye - y)**2)
```

the 's2' equation I use

```python
s2_eqn = (xe - x)**2 + (ye - y)**2
```

```python
s2_eqn
```

subbing in ye(xe) to s and s2

```python
s_eqn_2 = s_eqn.subs(ye,ye_xe_eqn)
s2_eqn_2 = s2_eqn.subs(ye,ye_xe_eqn)
```

```python
s_eqn_2
```

```python
s2_eqn_2
```

compute derivatives

```python
s2_eqn_2 = s2_eqn_2.expand()
```

```python
s2_eqn_2
```

```python
#ds_dxe_eqn = sp.diff(s_eqn_2,xe) doesn't look as nice
ds2_dxe_eqn = sp.diff(s2_eqn_2,xe)
```

```python
ds2_dxe_eqn
```

```python
ds2_dxe_eqn.args[0]
```

```python
ds2_dxe_eqn.args[1]
```

```python
ds2_dxe_eqn.args[2]
```

```python
lhs = (-ds2_dxe_eqn.args[0]-ds2_dxe_eqn.args[1]-ds2_dxe_eqn.args[2])**2
rhs = ds2_dxe_eqn.args[3]**2
ds2_dxe_eqn_step2 = rhs-lhs
```

```python
ds2_dxe_eqn_step3 = sp.simplify(ds2_dxe_eqn_step2*(a**4-a**2*xe**2)).expand(basic=True)
```

```python
ds2_dxe_eqn_step3_coeffs = sp.Poly(ds2_dxe_eqn_step3,xe).coeffs()
```

```python
ds2_dxe_eqn_step3_coeffs
```

```python
out_dseq0 = sp.solve(ds2_dxe_eqn_step3,xe)
```
